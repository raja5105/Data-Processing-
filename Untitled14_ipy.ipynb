```python
import pandas as pd
import dask.dataframe as dd
from dask.distributed import Client

# Start a Dask client
client = Client()

# Read the datasets using pandas
df_full = pd.read_csv("/content/Trips_Full Data.csv")
df = pd.read_csv("/content/Trips_by_Distance.csv")

# Data Cleaning with pandas
# Fill NaN values with 0 for 'Population Staying at Home'
df["Population Staying at Home"] = df["Population Staying at Home"].fillna(0)

# Convert 'Population Staying at Home' to integer
df["Population Staying at Home"] = df["Population Staying at Home"].astype("int64")

# Display column data types
print("Data Types:")
print(df.dtypes)

# Count non-null values in each column
print("\nCount Non-Null:")
print(df.notnull().sum())

# Count null values in each column
print("\nCount Null:")
print(df.isna().sum())

# Convert the cleaned pandas DataFrame to a Dask DataFrame for parallel computation
df_dask = dd.from_pandas(df, npartitions=4)

# Compute descriptive statistics for the whole dataframe
print("\nDescriptive Statistics for the whole DataFrame:")
print(df_dask.describe().compute())

# Compute descriptive statistics for just the 'Population Staying at Home' column
print("\nDescriptive Statistics for 'Population Staying at Home':")
print(df_dask['Population Staying at Home'].describe().compute())

# Shut down the Dask client
client.shutdown()
```
